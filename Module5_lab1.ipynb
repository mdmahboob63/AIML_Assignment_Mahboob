We have seen a graph of a linear regression model attempting to represent polynomial data, and how it falls short when trying to model non-linear data. Now, what if we took some non-linear noisy data, and fit a very high degree polynomial to it? Try to guess what would happen!

```python
import numpy as np
from sklearn import linear_model,metrics
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
import math
import matplotlib.pyplot as plt
```

```python
# POLYNOMIAL REGRESSION

def poly_regression(x,y,deg):

  x = x.reshape((-1,1))
  y = y.reshape((-1,1))

  poly = PolynomialFeatures(degree=deg)

  x_ = poly.fit_transform(x)
  poly.fit(x_,y)

  m1 = linear_model.LinearRegression()
  m1.fit(x_,y)
  y_pred = m1.predict(x_)

  mse = np.sqrt(metrics.mean_squared_error(y_pred,y))

  plt.plot(x,y_pred,color="red")
  plt.title("Visualization of actual data vs predicted outputs (polynomial regression)")
  plt.scatter(x,y,color="blue")
  plt.legend(["predicted polynomial","actual data points"])
  plt.show()

  print("The RMSE for polynomial regression is: {}".format(mse))
```

```python
deg = 2#@param {type:"slider", min: 1, max:30, step:1}

np.random.seed(10)

x_ex = np.array([i*np.pi/180 for i in range(60,300,6)])
y_ex = np.sin(x_ex) + np.random.normal(0,0.15,len(x_ex))
poly_regression(x_ex,y_ex,deg)
```

Now our observation is to check when underfitting and overfitting happens.

If we take low degree,the regression line is not fitting with the data point properly and covering very less points.This is called Underfitting.This is done usually when ML models not trained properly.It leads to poor predictions with test data.
If we take high degree,the regression line is trying to fit maximum datapoints or data samples also including unnecessary and noisy data. This is called overfitting.This is done usually when ML model is trained on more data including noisy also.as a result,an overfitted model may perform very well on the training data but poorly on new ,unseen data because it aligns with training data instead of generalizing on new data or patterns.
